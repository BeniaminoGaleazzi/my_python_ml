```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.arange(2,21)
y = 2*x +5
```

```python
print(x)
```

```python
print(y)
```

```python
# Create an object
plt.figure(figsize=(9,4))
# Plot function
plt.plot(x,y,label = "My data")
# title
plt.title('First Plot')

#X-Y labels
plt.xlabel('Values of x')
plt.ylabel('Values of y')

#Legend
plt.legend()
plt.show()
```

```python
# Create an object
plt.figure(figsize=(9,4))
# Plot function
plt.plot(x,y,'gd',label = "My data")
# title
plt.title('First Plot')

#X-Y labels
plt.xlabel('Values of x')
plt.ylabel('Values of y')

#Legend
plt.legend()
plt.show()
```

```python
#Plot the exponential
x = np.arange(1,11)
y = np.exp(x)
```

```python
plt.figure(figsize=(8,5))

# plot function

plt.title("Second plot")
plt.xlabel("Xlabel")
plt.ylabel("Ylabel")
plt.plot(x,y,'gd', label="My second Data")
plt.legend()
plt.grid()
plt.show()
```

```python
#data
x = np.arange(1,11)
y1 = x**2
y2 = x**3
y3 = x**4
```

```python
plt.figure(figsize=(8,5))
plt.plot(x, y1, 'g--', label="Squared power")
plt.plot(x, y2, 'bo', label="Third power")
plt.plot(x, y3, 'r^', label="Fourth power")

plt.show()
```
